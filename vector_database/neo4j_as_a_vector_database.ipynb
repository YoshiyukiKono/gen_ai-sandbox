````python
import os
img_paths = []
for root, dirs, files in os.walk("deepface/tests/dataset/"):
    for file in files:
        if '.jpg' in file:
            img_path = root+file
            img_paths.append(img_path)
````

````python
img_paths
````

````python
instances = {}
for img_path in img_paths:
    #raw image name without path and extension
    label = img_path.split("/")[-1].split(".")[0]

    embedding = DeepFace.represent(img_path=img_path, model_name = "Facenet")[0]["embedding"]

    instances[label] = embedding
````

````python
instances
````

````python
from neo4j import GraphDatabase, basic_auth

#SCHEMA = "neo4j+s"
#URI = SCHEMA + "://" + CONNECTION_URL
driver = GraphDatabase.driver(CONNECTION_URI, auth=basic_auth("neo4j", PASSWORD))

session = driver.session()
````

````python
from neo4j import GraphDatabase

# URI examples: "neo4j://localhost", "neo4j+s://xxx.databases.neo4j.io"
URI = "<URI for Neo4j database>"
AUTH = ("<Username>", "<Password>")

with GraphDatabase.driver(URI, auth=AUTH) as driver:
    driver.verify_connectivity()
````

````python
#flush database
result = session.run("MATCH (n) DETACH DELETE n")
#MATCH (n {name: 'img1'}) DETACH DELETE n
````

````python
statements = ""
for img_label in instances.keys():
    statements += "MERGE (%s_face:Face {name:'%s'})" \
                  " SET %s_face.embedding = %s \n" % (img_label
                  , img_label, img_label, instances[img_label])
````

````python
statements
````

````python
with session.begin_transaction() as trx:
    trx.run(statements)
    trx.commit()
````

```
MATCH (p1:Face)
MATCH (p2:Face)
WHERE p1.name <> p2.name
WITH p1, p2
, gds.similarity.euclidean(p1.embedding, p2.embedding) as distance
WHERE distance < 10
MERGE (p1)-[e:distance]-(p2)
SET e.distance=distance
```
